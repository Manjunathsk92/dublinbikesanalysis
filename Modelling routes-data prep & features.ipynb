#### Finding homologous routes

```python
def findHomologousRoutes(routes_list, route_prefix):
    homologous_routes=[]
    for route in routes_list:
        for i in route_prefix:
            if route not in homologous_routes and route.startswith(i):
                homologous_routes.append(route)
    return homologous_routes
```

```python
stops_coords = json.loads(open('/home/student/dbanalysis/stops_trimmed.json','r').read())
```

```python
#turn the warnign off lol
pd.options.mode.chained_assignment = None 
```

```python
#appending from stop coords and name
to_concat=[]
for stop in df['fromstop'].unique():
    
    temp_df=df[df['fromstop']==stop]
    if str(stop) in stops_coords:
        temp_df['from_stop_lat']=stops_coords[str(stop)]['lat']
        temp_df['from_stop_lon']=stops_coords[str(stop)]['lon']
        temp_df['from_stop_name']=stops_coords[str(stop)]['stop_name']
        to_concat.append(temp_df)
df=pd.concat(to_concat, axis=0)
```

```python
#appending to stop coords and name
to_concat=[]
for stop in df['tostop'].unique():
    
    temp_df=df[df['tostop']==stop]
    if str(stop) in stops_coords:
        temp_df['to_stop_lat']=stops_coords[str(stop)]['lat']
        temp_df['to_stop_lon']=stops_coords[str(stop)]['lon']
        temp_df['to_stop_name']=stops_coords[str(stop)]['stop_name']
        to_concat.append(temp_df)
df=pd.concat(to_concat, axis=0)
```

```python
from geopy.distance import vincenty

def pandasVincenty(row):
    '''calculate distance (m) between two lat&long points using the Vincenty formula '''

    return vincenty((row.from_stop_lat, row.from_stop_lon), (row.to_stop_lat, row.to_stop_lat)).meters
```

```python
df['distance'] =  df.apply(lambda r: pandasVincenty(r), axis=1)
```
