## Mapping routes, trips and individual trip ids.

```python
Rmap={}
for route_id in stops_trips['route_id'].unique():
    df = stops_trips[stops_trips['route_id']==route_id]
    Rmap[route_id]={}
    for shape_id in df['shape_id'].unique():
        gf = df[df['shape_id']==shape_id]
        
        Rmap[route_id][shape_id]={'trips':[trip for trip in gf['trip_id'].unique()]}
        Rmap[route_id][shape_id]['uniquetrips']=[]
        Rmap[route_id][shape_id]['uniquestops']=[]
        for trip_id in gf['trip_id'].unique():
            hf = gf[gf['trip_id']==trip_id].sort_values(by=['stop_sequence'])
            stps = tuple(stp for stp in hf['stop_id'])
            if stps not in Rmap[route_id][shape_id]['uniquestops']:
                Rmap[route_id][shape_id]['uniquestops'].append(stps)
                Rmap[route_id][shape_id]['uniquetrips'].append(trip_id)
        
        
```
