```python
def prep_test_stop(filename,weather,fromstop,tostop):
    
    df=pd.read_csv(filename,names=hds.get_stop_link_headers())
    df['fromstop']=fromstop
    df['tostop']=tostop
    df['traveltime']=df['actualtime_arr_to']-df['actualtime_dep_from']
    df['dwelltime']=df['actualtime_dep_from']-df['actualtime_arr_from']
    df['distance'] = s_getter.get_stop_distance(fromstop,tostop)
    df['speed'] = df['distance'] / (df['traveltime']/3600)

    df['date']=pd.to_datetime(df['dayofservice'],format= "%d-%b-%y %H:%M:%S").dt.date
    df['hour']=df['actualtime_arr_from']//3600
    df = pd.merge(df,weather, on=['date','hour'])
    return df
```

# Testing models across all stops

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn import metrics
```

```python
print('hello')
```

```python
def random_test_one_features():
    df=random_stop_data()
    msk = np.random.rand(len(df)) < 0.8
    train = df[msk]
    test=df[~msk]
    clf = LinearRegression(fit_intercept=True).fit(train[['actualtime_dep_from']],train['traveltime'])
    preds = clf.predict(test[['actualtime_dep_from']])
    return metrics.mean_absolute_error(test['traveltime'],preds),metrics.mean_squared_error(test['traveltime'],preds)
    
    
```

```python
err1=0
err2=0
total=0
for i in range(1000):
    try:
        err0,err01=random_test_one_features()
        err1+=err0;err2+=err01
        total+=1
    except:
        pass
print(err1/total,(err2**0.5)/total)
```

```python
def random_test_two_features():
    df=random_stop_data()
    df['day']=df['dt'].dt.day
    msk = np.random.rand(len(df)) < 0.8
    train = df[msk]
    test=df[~msk]
    clf = LinearRegression(fit_intercept=True).fit(train[['actualtime_dep_from','day']],train['traveltime'])
    preds = clf.predict(test[['actualtime_dep_from','day']])
    del df
    return metrics.mean_absolute_error(test['traveltime'],preds),metrics.mean_squared_error(test['traveltime'],preds)
```

```python
err1=0
err2=0
total=0
for i in range(100):
    try:
        err0,err01=random_test_two_features()
        err1+=err0;err2+=err01
        total+=1
    except:
        pass
print(err1/total,(err2**0.5)/total)
```

```python
def random_test_three_features():
    df=random_stop_data()
    df['day']=df['dt'].dt.day
    msk = np.random.rand(len(df)) < 0.8
    train = df[msk]
    test=df[~msk]
    clf = LinearRegression(fit_intercept=True).fit(train[['actualtime_dep_from','day','rain']],train['traveltime'])
    preds = clf.predict(test[['actualtime_dep_from','day','rain']])
    del df
    return metrics.mean_absolute_error(test['traveltime'],preds),metrics.mean_squared_error(test['traveltime'],preds)
```

```python
err1=0
err2=0
total=0
for i in range(1000):
    try:
        err0,err01=random_test_three_features()
        err1+=err0;err2+=err01
        total+=1
    except:
        pass
print(err1/total,(err2**0.5)/total)
```
