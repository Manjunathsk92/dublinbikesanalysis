```python
import pandas as pd
import os
os.listdir('/home/student/dbanalysis/dbanalysis/resources')
```

```python
df = pd.read_csv('/home/student/dbanalysis/dbanalysis/resources/cluster_frame.csv')
```

```python
df=df.fillna(0)
import numpy as np
df=df.replace([np.inf, -np.inf], np.nan)
df=df.fillna(0)
```

```python
cols = [col for col in df.columns if col not in ['distance','stopA','stopB'] and col[:5]!='dwell']
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=100).fit(df[cols])
```

```python
clustered = df
clustered['labels']=kmeans.labels_
```

```python
cluster50 = clustered[clustered['labels']==2]
```

```python
cluster50
```

## Let's try and build a model for this cluster then

```python
print('hello')
```

```python
from dbanalysis import stop_tools
s_getter = stop_tools.stop_getter()
to_concat = []
for row in cluster50.itertuples():
    stopA = str(row[1])
    stopB = str(row[2])
    df = stop_tools.get_stop_link(stopA,stopB,merge_weather=True)
    df['distance']=s_getter.get_stop_distance(stopA,stopB)
    df['stopA']==stopA
    df['stopB']==stopB
    to_concat.append(df)
df=pd.concat(to_concat,axis=0)
```

```python
len(df)
```

```python
df['traveltime'].mean()
```

```python
df=df.dropna()
```

```python
df['traveltime']=df['actualtime_arr_to']-df['actualtime_dep_from']

time_format="%d-%b-%y %H:%M:%S"
df['dt']=pd.to_datetime(df['dayofservice'],format=time_format)
df['day']=df['dt'].dt.dayofweek
df['month']=df['dt'].dt.month
```

## With Random Forest Regressor

```python
features = ['actualtime_dep_from','distance','hour','month','rain','temp']
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# 80:20 split
import numpy as np
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
regr = RandomForestRegressor().fit(train[features],train['traveltime'])
```

```python
predicts = regr.predict(test[features])
```

```python
from sklearn import metrics
print(metrics.mean_absolute_error(test['traveltime'],predicts))
print(metrics.r2_score(test['traveltime'],predicts))
```

```python
#for single stop
test2 = test[(test['stopA']=='2014') & (test['stopB']=='2016')]
```
