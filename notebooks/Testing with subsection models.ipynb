```python
from dbanalysis.classes import subsection_model
```

```python
route15=routes['15'][3][1:]
```

```python
import pandas as pd
from dbanalysis import headers as hds
arrives = pd.read_csv('/home/student/data/stops/'+str(route15[0])+'/'+str(route15[1])+'.csv',names=hds.get_stop_link_headers())
```

```python
ends = pd.read_csv('/home/student/data/stops/'+str(route15[-2])+'/'+str(route15[-1])+'.csv',names=hds.get_stop_link_headers())
```

```python
nodes = []
for stop_id in route15:
    nodes.append(subsection_model.stop(stop_id,from_pickle=False,train_test='year',validate=True,model_dwell_time=False,dimension=3,regressor_type='Linear'))
```

```python
print(len(nodes))
```

```python
df = arrives['actualtime_arr_from']
```

```python
arrives['begin']=arrives['actualtime_arr_from']
ends['end']=ends['actualtime_arr_to']
```

```python
df=pd.merge(arrives[['begin','tripid','dayofservice']],ends[['end','tripid','dayofservice']],on=['tripid','dayofservice'])

```
