```python
def findHomologousRoutes(routes_list, route_prefix):
    homologous_routes=[]
    for route in routes_list:
        for i in route_prefix:
            if route not in homologous_routes and route.startswith(i):
                homologous_routes.append(route)
    return homologous_routes
```

```python
def merge_weather(weather_df,df):
    """Takes 2 dataframes- a route dataframe after being munged by routetools and historical_weather_data.csv - and merges them based on dt """
    #converting date column in weather table to datetime
    datetimes=pd.to_datetime(weather_df['date'])
    weather_df['dt']=datetimes
    #converting dep and arr times in df to time series 
    df['plannedtime_arr_from']=pd.to_timedelta(df['plannedtime_arr_from'], unit='s')
    df['plannedtime_dep_from']=pd.to_timedelta(df['plannedtime_dep_from'], unit='s')
    df['actualtime_arr_from']=pd.to_timedelta(df['actualtime_arr_from'], unit='s')
    df['actualtime_dep_from']=pd.to_timedelta(df['actualtime_dep_from'], unit='s')
    df['plannedtime_arr_to']=pd.to_timedelta(df['plannedtime_arr_to'], unit='s')
    df['actualtime_arr_to']=pd.to_timedelta(df['actualtime_arr_to'], unit='s')
    #datetimes1=pd.to_datetime(df['dayofservice'])
    #df['dt']=datetimes1
    weather_route = pd.merge(weather_df, df, on=df['actualtime_arr_from'])
    #weather_route.drop(['date'],axis=1)
    return weather_route    
```
